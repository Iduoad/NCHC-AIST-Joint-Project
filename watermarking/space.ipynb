```python
import reversible
import time
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
```

```python
p = (4, 4, 4)
a = (3, 3, 3)
b = (1, 1, 1)

print("origin pixel\t: %s" % (p,))
q = reversible.F_inverse(b, reversible.F(a, p))
print("watermark pixel\t: %s" % (q,))
p_prime = reversible.F_inverse(a, reversible.F(b, q))
print("recovered pixel\t: %s" % (p_prime,))
```

```python
Image.MAX_IMAGE_PIXELS = 1000000000
pil_im = Image.open('/tmp/space.tiff', 'r')
imshow(np.asarray(pil_im))
```

```python
logo = Image.open('./logo_2.jpg', 'r').convert('RGB')
imshow(np.asarray(logo))
```

```python
start1 = time.time()
watermarking_img = reversible.addWatermark("/tmp/space.tiff", "./logo_2.jpg", blk_height=3000, blk_width=6000)
end1 = time.time()
print(end1 - start1)

imshow(np.asarray(watermarking_img))
```

```python
start2 = time.time()
recovered_img = reversible.removeWatermark(watermarking_img, "./logo_2.jpg", blk_height=3000, blk_width=6000)
end2 = time.time()
print(end2 - start2)

imshow(np.asarray(recovered_img))
```
