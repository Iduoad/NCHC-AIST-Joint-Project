```python
import reversible
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
```

```python
p = (4, 4, 4)
a = (3, 3, 3)
b = (1, 1, 1)

print("origin pixel\t: %s" % (p,))
q = reversible.F_inverse(b, reversible.F(a, p))
print("watermark pixel\t: %s" % (q,))
p_prime = reversible.F_inverse(a, reversible.F(b, q))
print("recovered pixel\t: %s" % (p_prime,))
```

```python
pil_im = Image.open('./aa.jpg', 'r')
imshow(np.asarray(pil_im))
```

```python
logo = Image.open('./logo.png', 'r').convert('RGB')
imshow(np.asarray(logo))
```

```python
estimate_A, watermarking_img = reversible.addWatermark("./aa.jpg", "./logo.png")
imshow(np.asarray(watermarking_img))
```

```python
recovered_img = reversible.removeWatermark(watermarking_img, estimate_A, "./logo.png")
imshow(np.asarray(recovered_img))
```
